### Two options for sublots

```python
#subplot option1
fig, ((ax1 , ax2) , (ax3 , ax4)) = plt.subplots(nrows=2,
                                                ncols=2,
                                                figsize=(10,5))

#plot to each differentn axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10) , np.random.random(10))
ax3.bar(nut_butter_price.keys() , nut_butter_price.values())
ax4.hist(np.random.randn(1000));
```

```python
#subplot option2 
fig , ax = plt.subplots(nrows=2,
                        ncols=2,
                        figsize=(10,5))

ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10) , np.random.random(10))
ax[1,0].bar(nut_butter_price.keys() , nut_butter_price.values())
ax[1,1].hist(np.random.randn(1000));
```

## Which One should you use? (Pyplot VS matplotlib OO method)
* When plotting something quickly , Okay to use pyplot method 
* When plotting something more advanced , use the OO method

```python
heart_disease.head()
```

```python
len(heart_disease)
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#we want to plot scatter by 'age' with 'chol' and colored by target
#1st : Pyplot method : plot directly from pandas dataFrame with plot
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             figsize=(10,6));
```

```python
#2nd : Object Oriented method
fig, ax = plt.subplots(figsize=(10,6));
over_50.plot(kind="scatter",
                       x="age",
                       y="chol",
                       c="target",
                       ax=ax);

#ax.set_xlim([50 , 60])
```

```python
over_50["target"].values
```

```python
#3rd : OO method from scratch***
fig , ax = plt.subplots( figsize=(10,6) )

#plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])


#customize the plot 
ax.set(title="Heart disease and cholesterol Levels",
       xlabel="age",
       ylabel="cholesterol");
       
#Add a legend
ax.legend(*scatter.legend_elements() , title="target");

#create a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
#subplot of chol, age, thalach
fig , (ax0,ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)
#Add some data to ax0----------------------------------------
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])
#customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="cholesterol")
#legend
ax0.legend(*scatter.legend_elements() , title="target")
#add a meanLine
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");


#Add some data to ax1-----------------------------------------
scatter2 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"])
# customize
ax1.set(title="Heart Disease and Max_heart_Rate(thalach)",
        xlabel="age",
        ylabel="thalach")
# legend
ax1.legend(*scatter2.legend_elements() , title="target")
# horizanal line
ax1.axhline(over_50["thalach"].mean(),
            linestyle="--")

#FIGURE
fig.suptitle("Heart Disease Analysis", fontsize="16", fontweight="bold");
```
